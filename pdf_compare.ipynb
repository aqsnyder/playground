```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

def compare_pdfs(pdf1_path, pdf2_path):
    """Compares the text content of two PDFs and prints differences."""
    text1 = extract_text_from_pdf(pdf1_path)
    text2 = extract_text_from_pdf(pdf2_path)

    # Compare the two texts line by line
    text1_lines = text1.splitlines()
    text2_lines = text2.splitlines()

    differences = []

    for i, (line1, line2) in enumerate(zip(text1_lines, text2_lines), 1):
        if line1 != line2:
            differences.append(f"Difference in line {i}:\nPDF1: {line1}\nPDF2: {line2}\n")

    if not differences:
        print("The PDF files are identical.")
    else:
        print("Differences found:")
        for diff in differences:
            print(diff)

# Example usage:
pdf1_path = "pdf_files/Aaron Snyder Employment Contract_signed.pdf"
pdf2_path = "pdf_files/Aaron Snyder Employment Contract_original.pdf"

compare_pdfs(pdf1_path, pdf2_path)
```
